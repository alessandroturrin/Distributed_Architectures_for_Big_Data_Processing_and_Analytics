```python
from graphframes import GraphFrame

spark = SparkSession.builder.getOrCreate()

airlinesDF = spark.read.load('airlines.csv', format='csv', header=True, inferSchema=True)
airportsDF = spark.read.load('airports.csv', format='csv', header=True, inferSchema=True)
routesDF = spark.read.load('routes.csv', format='csv', header=True, inferSchema=True)
```

```python
v = airportsDF.withColumn("id", airportsDF.id.cast("string"))
e = routesDF.select("airport_source_id","airport_destination_id")\
          .withColumn("airport_source_id", routesDF.airport_source_id.cast("string"))\
          .withColumn("airport_destination_id", routesDF.airport_destination_id.cast("string"))\
          .withColumnRenamed("airport_source_id","src")\
          .withColumnRenamed("airport_destination_id","dst")

g = GraphFrame(v,e).cache()
```
